```python
import pandas as  pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')




from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MultiLabelBinarizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier

from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import time
```

```python
file_path = r'A:\final year project\AyuCare\cbc_data.csv'    
df = pd.read_csv(file_path)
```

```python
from sklearn.preprocessing import LabelEncoder
X = df.drop(columns=['Disease'],axis=1)
y = df['Disease']

le=LabelEncoder()
le.fit(y)
y=le.transform(y)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
#Logistic Regression
log_reg_model = OneVsRestClassifier(LogisticRegression(max_iter=300, solver='lbfgs'))
log_reg_model.fit(X_train, y_train)

y_pred_log_reg = log_reg_model.predict(X_test)
log_reg_accuracy = accuracy_score(y_test, y_pred_log_reg)


model_performance = {}
model_performance['Logistic Regression'] = log_reg_accuracy

print("Logistic Regression Accuracy:", log_reg_accuracy)
```

```python
# Decision Tree
dt_model = DecisionTreeClassifier(max_depth=20, random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, y_pred_dt)
model_performance['Decision Tree'] = dt_accuracy

```

```python
from sklearn.naive_bayes import GaussianNB

# Naive Bayes (GaussianNB)
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)
nb_accuracy = accuracy_score(y_test, y_pred_nb)
model_performance['Naive Bayes'] = nb_accuracy
```

```python
rf_model = RandomForestClassifier(random_state=42, n_jobs=-1, n_estimators=10)  # Parallelize and use 100 trees
rf_model.fit(X_train, y_train)  # Train with the encoded labels

# Predict and calculate accuracy
y_pred_rf = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")
print("Model Performance Keys:", model_performance.keys())
model_performance['Random Forest'] = rf_accuracy
#print(f"Random Forest Accuracy: {rf_accuracy}")
```

```python

print("Model Performance:")
for model_name, accuracy in model_performance.items():
    print(f"{model_name}: {accuracy:.4f}")
```
